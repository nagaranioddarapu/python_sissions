```python
# extract the name list,surname list,company list,email id list
file_path=r'emails.txt'
file=open(file_path,encoding='utf-8-sig')
b = file.read()
res = b.split(';')
l1=[]
l2=[]
l3=[]
d1={}

l1=[i[0:i.find(',')] for i in res]
l2=[i[i.find(' ')+1:i.find(' ',i.find(' ')+1)] for i in res]
l3=[i[i.find('<'):i.find('>')] for i in res]
l4=[i[i.find('@')+1:i.find('.com')] for i in res]
print(l4)

d1['first_name']=l1
d1['second_name']=l2
d1['mailid']=l3
d1

   
```
